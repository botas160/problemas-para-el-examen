```
while 1: 
  a=150
  b=170
  c=10000

  pay=int(input("Cuantos payasos contiene su pedido "))
  mes=int(input("Cuantos juegos de mesa contiene su pedido "))
  ppay=pay*a
  pmesa=mes*b
  ptotal=ppay+pmesa
  if ptotal>c:
    print("False")
    residuo=ptotal-c
    print("El excedente del pedido es ",residuo)

  else: 
    print("Gracias por su compra ")
```
